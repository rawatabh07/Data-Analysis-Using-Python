```python
coorelation_matrix=df.corr(method='pearson')
sns.heatmap(coorelation_matrix,annot=True)
plt.title('Coorelation matric for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie features')
plt.show()
```

```python
#lets set categoery for each column
df_num = df
for col_name in df_num.columns:
    if(df_num[col_name].dtype == 'object'):
        df_num[col_name]= df[col_name].astype('category')
        df_num[col_name] = df[col_name].cat.codes
df_num
```

```python
coorelation_matrix=df_num.corr(method='pearson')
sns.heatmap(coorelation_matrix,annot=True)
plt.title('Coorelation matric for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie features')
plt.show()
```

```python
df_num.corr()
```

```python
corr_mat = df_num.corr()
corr_pairs = corr_mat.unstack()
print(corr_pairs)
```

```python
sort_pairs = corr_pairs.sort_values(kind="quicksort")
print(sort_pairs)
```

```python
high_cor = sort_pairs[abs(sort_pairs)>0.5]
high_cor
```
